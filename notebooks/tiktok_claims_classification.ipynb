### **Task 3: Feature engineering**

Extract the length of each `video_transcription_text` and add this as a column to the dataframe, so that it can be used as a potential feature in the model.

```python
# Extract the length of each `video_transcription_text` and add this as a column to the dataframe
data["video_transcription_length"] = data["video_transcription_text"].str.len()
```

Calculate the average text_length for claims and opinions.

```python
# Calculate the average text_length for claims and opinions
avg_claim_length = data[data['claim_status'] == 'claim']['video_transcription_length'].mean()
avg_opinion_length = data[data['claim_status'] == 'opinion']['video_transcription_length'].mean()
```

Visualize the distribution of `text_length` for claims and opinions.

```python
# Visualize the distribution of `text_length` for claims and opinions
# Create two histograms in one plot
plt.figure(figsize=(10,5))
plt.hist(data[data['claim_status'] == 'claim']['video_transcription_length'], bins=30, alpha=0.5, label='Claim')
plt.hist(data[data['claim_status'] == 'opinion']['video_transcription_length'], bins=30, alpha=0.5, label='Opinion')
plt.axvline(avg_claim_length, color='blue', linestyle='dashed', linewidth=1, label='Avg Claim Length')
plt.axvline(avg_opinion_length, color='orange', linestyle='dashed', linewidth=1, label='Avg Opinion Length')
plt.title('Distribution of Video Transcription Lengths')
plt.xlabel('Video Transcription Length')
plt.ylabel('Frequency')
plt.legend()
plt.show()      
```
